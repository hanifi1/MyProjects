```python
import requests   #import LIBRARIES
import pandas as pd

import re
clean = re.compile('<.*?>') #this is the line for clean text 

from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

df = pd.read_csv('D:/Machine Learnaing/Web_Scrapping/df.csv')
df.head()
```

```python

def read_link(link):  #function to read a text and coonvert to soup
    req = Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html, 'lxml')
    return soup

def read_title(soup):   #read the soup and extract the title
    ti = soup.title
    clean_ti = str(ti)[7:-8]
    return clean_ti


def texts_list(soup):   #read the soup and clean the body of text. output is a List 
    texts = soup.find_all('p')
    text_list = []
    for text in texts:
        clean_text = (re.sub(clean, '',str(text)))
        text_list.append(clean_text)
    return text_list

def text_date(List):  #read the list and finding date
    if List[6] == 'News provided by':
        t = List[7]
    else:
        t = List[6]
    return t
```

```python
for i in range(657, len(df)):
    Link = df.loc[i, 'URL']
    print(Link)
    Soup = read_link(Link)
    df.loc[i, 'title'] = read_title(Soup)
    List = texts_list(Soup)
    Date = text_date(List)
    df.loc[i, 'Date'] = Date
    print(Date)         # test progres
    name = df.loc[i, 'URL'][-14: -5] + '.txt'
    df.loc[i, 'text_file'] = name
    print(name)
    print(i)
    try: 
        try:
            main = List[8:-12]
            f = open(name, 'w')
            for t in main:
                f.write(t + '\n')
            f.close()
        except UnicodeEncodeError:
            main = List[8:-14]
            f = open(name, 'w',encoding='utf-8')
            for t in main:
                f.write(t + '\n')
            f.close()
    except HTTPError:
        pass
```
