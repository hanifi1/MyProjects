```python
import requests
import bs4
from urllib.request import urlopen
import pandas as pd
```

```python
URLS = ['https://ca.indeed.com/jobs?q=finance+analyst&l=Montreal+QC']
count = ['10', '20', '30', '40', '50', '60', '70', '80']

for i in count:
    add = URLS[0] + '&start' + i
    URLS.append(add)
    
URLS
```

```python
Links = []
root = 'https://ca.indeed.com'
for URL in URLS:
    html = urlopen(URL)
    soup = bs4.BeautifulSoup(html)
    all_links = soup.find_all('a')
    for link in all_links:
        ad = link.get('href')
        if ad != None:
            if ad.find('pagead') > 0:
                ad = root + ad
                Links.append(ad)
    
```

```python
df = pd.DataFrame(data={'Financial Analyst Jobs': Links})
df.head()
```

```python
df.to_csv('indeed.csv', index=False)
```
